```python
import pandas as pd
base_url = "https://www.tripadvisor.com/Restaurants-g293825-Abuja_Federal_Capital_Territory.html"
urls = [base_url if i == 0 else f"https://www.tripadvisor.com/Restaurants-g293825-oa{i}-Abuja_Federal_Capital_Territory.html" for i in range(0, 359, 30)]
import requests
from bs4 import BeautifulSoup
import re
from fake_useragent import UserAgent

def get_restaurant_links(url):
    ua = UserAgent()
    headers = {'User-Agent': ua.random}
    # Set a user-agent to mimic a browser
    
    try:
        # Send GET request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors
        
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # TripAdvisor typically has restaurant links in elements with specific patterns
        # Look for links containing "Restaurant_Review" which is common in TripAdvisor restaurant URLs
        restaurant_links = []
        
        # Find all <a> tags that might contain restaurant links
        links = soup.find_all('a', href=re.compile(r'Restaurant_Review'))
        
        # Extract the full URLs
        base_url = "https://www.tripadvisor.com"
        for link in links:
            href = link.get('href')
            if href and 'Restaurant_Review' in href:
                # Create absolute URL if it's a relative link
                if href.startswith('/'):
                    full_url = base_url + href
                else:
                    full_url = href
                
                # Avoid duplicates
                if full_url not in restaurant_links:
                    restaurant_links.append(full_url)
        filtered_list = [s for s in restaurant_links if "#REVIEWS" not in s]
        return filtered_list
    
    except requests.exceptions.RequestException as e:
        print(f"Error fetching the URL: {e}")
        return []



# Get restaurant links
restaurant_links = []

for i in urls:
    restaurant_links.append(get_restaurant_links(i))
    print(i)

from itertools import chain
eat_links  = list(chain(*restaurant_links))
eat_list = list(set(eat_links))
    
```

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import time
# URL of the TripAdvisor page

def get_restaurant_info(url):
    # Set up fake headers
    ua = UserAgent()
    headers = {'User-Agent': ua.random}
    
    # Send request
    response = requests.get(url, headers=headers)
    time.sleep(3)
    soup = BeautifulSoup(response.text, 'html.parser')

# Scraping data

    data = {}

    # Name
    name_tag = soup.find("span", class_="ReGgO")
    data['name'] = name_tag.text.strip() if name_tag else "N/A"

    # Location
    location_tag = soup.find("a", href=lambda x: x and "/LocationPhoto" in x)
    if not location_tag:
        location_tag = soup.find("span", class_="yEWoV")  # backup
    data['location'] = location_tag.text.strip() if location_tag else "N/A"

    # Find a <span> tag that contains AM or PM in its text
    open_time_tag = soup.find("span", string=lambda text: text and ("AM" in text or "PM" in text))
    
    # Clean the text if found
    if open_time_tag:
        open_times = open_time_tag.get_text(strip=True)
    else:
        open_times = "N/A"
    
    data['open_times'] = open_times



    # Target the span with both classes
    meal_tag = soup.find("span", class_="alXOW EEXWj")
    
    # Get the text if the tag is found
    if meal_tag:
        meals = meal_tag.get_text(strip=True)
    else:
        meals = "N/A"
    
    data['amenities'] = meals



    # Phone number
    phone_tag = soup.find("a", href=lambda x: x and "tel:" in x)
    data['phone'] = phone_tag.text.strip() if phone_tag else "N/A"

    # Website
    website_tag = soup.find("a", href=lambda x: x and "http" in x and "tripadvisor.com" not in x and "tripadvisorsupport.com" not in x)
    data['website'] = website_tag['href'].strip() if website_tag else "N/A"

    # Rating
    rating_tag = soup.find("div", class_="VVbkp")
    data['rating'] = rating_tag.text.strip() if rating_tag else "N/A"
    data['url'] = url

    return data


def scrape_multiple_restaurants(url_list):
    all_data = []
    for url in url_list:
        try:
            info = get_restaurant_info(url)
            all_data.append(info)
        except Exception as e:
            print(f"Error scraping {url}: {e}")
    return pd.DataFrame(all_data)




df = scrape_multiple_restaurants(eat_list)


```
